### Fungsi

```
import math

def luas_lingkaran(r: int=2):
  """Menghitung luas lingkaran dengan jari-jari r menggunakan rumus A = phi * r^2"""
  area = 3.14 * (r**2)
  return area
print(f"area: {luas_lingkaran():.2f}")

```
